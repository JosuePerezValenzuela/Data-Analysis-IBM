```python
'''
Como tratar con datos faltantes
1. Verficar si se puede conseguir las informacion faltante
2. Eliminar los datos faltantes
    2.1 Eliminar la variable
    2.2 Eliminar el registro
3. Reemplazar los datos faltantes
    3.1 Reemplazar con el promedio
    3.2 Reemplazarlo por lo mas comun
    3,3 Reemplazarlo con otras funciones
4. Dejarlo con datos faltantes

axis = 0 elimina la fila entera
axis = 1 elimina la columna entera
inplace = True Nos permite guardar en el mismo df al modificacion, sin este, tendriamos
que proporcionar otra variable
'''
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

Data normalization
Tenemos datos que estan dentro de un rango (0-120) y otro dato que esta en rangos de
(100000 - 1000000), entonces este ultimo influira mas que el otro dato
Entonces podemos normalizar los valores para que esten en un rango simular

Se tiene diferentes formas de normalizacion  
Simple feature scaling -> Valor / Xmax este divide cada valor entre el valor maximo de esa columna/feature  
Min-Max -> (Valor - Xmin) / (Xmax - Xmin) de esa columna  
Z-score -> (Valor - u) / o donde u es el promedio y o es la desviacion estandar

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max() 

df[["length","width","height"]].head()
```

```python
#ANOVA (Analysis of variance)
# Encontrar la correlacion entre diferentes variables categoricas (Grupos)
# Devuelve F-test score y p-value
# F-test calcular el ratio de variacion entre los promedios de los grupos
# p-value muestra si el resultado obtenido es estadisticamente significante
# F-test score alto indica que hay una diferencia significativa entre los grupos
# p-value bajo indica que la diferencia entre los grupos es estadisticamente significante
# si el p-value es menor a 0.05, se puede decir que hay una diferencia significativa 
# entre los grupos
# Si el f-test es alto y el p-value es bajo, hay una correlacion fuerte entre las
# variables categoricas y otras variables
from scipy import stats
df_anova = df[["make", "price"]]
grouped_anova=df_anova.groupby(["make"])
annova_results_1 = stats.f_oneway(grouped_anova.get_group("honda")["price"],
                                grouped_anova.get_group("subaru")["price"])
annova_results_1
```

```python
#Pearson correlation 
# Correlation coefficient (r) es una medida de la relacion lineal entre dos variables
# El valor de r puede variar entre -1 y 1
# Un valor de r cercano a 1 indica una correlacion positiva fuerte
# Un valor de r cercano a -1 indica una correlacion negativa fuerte
# Un valor de r cercano a 0 indica que no hay correlacion lineal entre las variables
# P-value es la probabilidad de que la correlacion observada sea debida al azar
# Un p-value bajo (< 0.05 o < 0.001) indica que la correlacion es estadisticamente significante
# Un p-value alto (> 0.1) indica que la correlacion no es estadisticamente significante
pearson_coef, p_value = stats.pearsonr(df["horsepower"], df["price"])
print("The Pearson correlation coefficient is", pearson_coef, "with a p-value of", p_value)
```
